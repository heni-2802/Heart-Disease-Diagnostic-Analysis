# HEART DISEASE DIAGNOSTIC ANALYSIS


```python
# Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Heart Disease Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Dropping rows with missing values
df = df.replace('?', np.nan)  
df = df.dropna()  
```

```python
# Convert to string 
df['sex'] = df['sex'].astype(str)
df['target'] = df['target'].astype(str)
```

```python
# Distribution of heart disease cases
sns.countplot(x='target', data=df)
plt.title('Heart Disease Distribution')
plt.show()
```

```python
# Heart disease by gender
sns.countplot(x='sex', hue='target', data=df)
plt.title('Heart Disease by Gender')
plt.show()
```

```python
# Heart disease by age
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='age', hue='target', multiple='stack', kde=True)
plt.title('Heart Disease by Age')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Interactive plot for heart disease by age
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', multiple='stack', bins=50, kde=True)
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Interactive scatter plot for cholesterol vs. max heart rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='chol', y='thalach', hue='target', palette='viridis')
plt.title('Cholesterol vs. Max Heart Rate by Heart Disease Status')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
# Interactive box plot for chest pain type vs. age
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='cp', y='age', hue='target', palette='Set2')
plt.title('Chest Pain Type vs. Age by Heart Disease Status')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.show()
```
